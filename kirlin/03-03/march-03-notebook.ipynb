# Loops

Our programs will often require that we write code that performs the same task multiple times. Duplicating code has a number of disadvantages, such as:
  - it's time consuming to copy-and-paste code over and over again, 
  - it makes your program needlessly long, 
  - and if you make a mistake, you'll need to fix it in multiple locations.

A **loop** is a concept in a programming language that allows you to repeat a section of code.  Most programming languages have a few kinds of loops.

## `while` loops

A **while loop** a loop that repeats a section of code *as long as (while) a condition is true.*

Python syntax:
<pre>
<font size=+1>
<font color=red><b>while</b></font> <i>condition</i>:
	statement              # These indented statements
	statement              # are called the body of
    more statements...     # the loop.
statement
statement
more statements...
</font>
</pre>
* The condition must be something that Python can determine to be __`True`__ or __`False`__.
* The indented group of statements is called the *body* of the loop.
* When the loop begins running, the *condition* is tested, and if it is `True`, all the statements in the *body* are executed. Otherwise, the *body* is skipped.  
* After the *body* is run, Python returns to the beginning of the loop and tests the *condition* again.  If 
the *condition* is still true, the *body* is run again.  This happens over and over (check condition, if true, run body, check condition, if true, run body...) until the condition becomes false.  At that point, the loop stops running.

### Compare and contrast this with an if statement:



````python
# Our first loop

keep_going = "yes"
while keep_going == "yes":
    print("I am inside a loop!")
    keep_going = input("Do you want this loop to keep going? ")
````

## More about while loops

- In order for a loop to stop executing, something has to happen inside the loop to make the condition `False`.
- The loop does not stop running instantaneously when the loop condition becomes `False`.  If the condition is `True`
when the when is tested, the entire body will execute.  
- Normally, something inside the body will change the loop condition from `True` to `False`, but when that happens,
the rest of the body will still run.


````python
# Our second loop

keep_going = "yes"
while keep_going == "yes":
    print("I am inside a loop!")
    keep_going = input("Do you want this loop to keep going? ")
    print("I am still inside the loop!")
````

### This is a common loop paradigm

The type of loop above is very common when you want the user to be allowed to have complete
control over whether the loop should run again or not:

```python
# Set up a variable to control the loop:
keep_going = "yes"

# Keep the loop going as long as the variable is 'yes'
while keep_going == "yes":

    # Do whatever you want here.  
    # Statements here will be repeated each time through the loop.

    # Ask the user if they want to keep going.
    keep_going = input("Do you want this loop to keep going? ")
```
