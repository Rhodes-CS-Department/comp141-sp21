# Graphics

* Many programming languages include a library for computer graphics 
* A __library__ is a pre-written collection of functions usually centered around a theme (graphics, networking, math, sound, etc.)
* Python provides many libraries: https://docs.python.org/3/library/
* Using libraries (rather than starting from scratch) enables you to write programs faster
* Libraries are centered around functions because we know functions are abstractions of algorithms.

## The canvas
* All graphics in Python are drawn in a separate area of the screen called the canvas.
* The canvas is laid out similarly to the Cartesian plane, but with a flipped y-axis.



## Your first graphics program

For more information about the functions included in our cs1.graphics library, please see the reference link (also linked to on Canvas):  https://rhodes.box.com/s/w5wpfrqg1g5olsoz45sy1djbah7cmwoe


````python
# Graphics demonstration

from cs1.graphics import *

def main():
    # Always be sure to open the canvas in main - you only want one canvas per program
    open_canvas(500, 500)
    
    # Draw a border around the canvas
    draw_rect(0, 0, 500, 500)
    
    # Draw a circle
    draw_circle(250, 250, 50)
  
  
main()
````

## Graphics and Colors
Colors in computer graphics are often represented by "RGB triples," which are three numbers, each
between 0 and 255.  The numbers represent the intensities of the colors red, green, and blue, which
are mixed to form one color.
```
0, 0, 0       #black
255, 255, 255 #white
255, 0, 0     #red
0, 255, 0     #green
0, 0, 255     #blue
255, 0, 255   #purple
```
