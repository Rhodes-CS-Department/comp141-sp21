## Abstraction in Programming: Functions

* A function is used to assign a name to a mini-program within a larger program.
* Any time you want to run the mini-program, you can use the name of the function instead of repeating the code for the mini-program.
* To use a function, we must define it first.

## What happens when a function is called?

* Python sees a line of code with a function call on it.
* Python jumps to the corresponding function definition and runs the statements in the body of the function.
* At the end of the function's body, Python jumps back to line of the program that originally called the function, and continues with the next line.

## The `main()` function

* This is the name of a function, that by convention in Python, is the first function that runs
when your program begins.
* This function is in charge of running the main logic of your program.


* From this point on in the class, always define a `main()` function in your programs.
* Always call the `main()` function as the last line in your program.




```python
# Example of defining and calling a main() function.

def twinkle():   
    print("Twinkle twinkle little star")
    print("How I wonder what you are")

def main():
    twinkle()  # Call the twinkle function
    print("Up above the world so high")
    print("Like a diamond in the sky")
    twinkle()  # Call the twinkle function again

main()  # Call the main function to start the program
```
