## How `if-elif-else` works:

* Python runs each condition in a group in order, top to
bottom.
* Once a condition is found that is True, the corresponding
statements are run, **and the rest of the conditions and
statements are ignored.**

```python
# Example:
x = 5
if x < 2:
    print("A")
elif x < 6:
    print("B")
elif x < 10:
    print("C")
```

Any new `if` creates a new group, independent of any previous `if`/`elif`/`else`:

```python
# Example:
x = 5
if x < 2:
    print("A")
elif x < 6:
    print("B")
if x < 10:
    print("C")
```

## Two more things

It is OK to mix `and`s and `or`s, but you should parenthesize
them:

<font size=+2 style="line-height: 1.5;">
<pre>
if <font color=red><b>(</b></font>x > 0 and y > 0<font color=red><b>)</b></font> or x < 0:
</pre>
   is different than
<pre>
if x > 0 and <font color=red><b>(</b></font>y > 0 or x < 0<font color=red><b>)</b></font>:
</pre>
</font>

If you ever want to **negate** a condition, you can use the word `not`:

<font size=+2 style="line-height: 1.5;">
<pre>
if grade >= 90:
    print("You get an A!")
</pre>
does the opposite of
<pre>
if <font color=red>not</font>(grade >= 90):
    print("You don't get an A.")
</pre>
</font>

`and`/`or`/`not` can be combined in any way, using parentheses to specify how they should be interpreted.
